```python
import pandas as panda
import numpy as numpy
import seaborn as seaborn
import matplotlib.pyplot as matplot
```

```python
# cargar el Dataset "real_state_ecuador_dataset.csv"
archivo = panda.read_csv("real_state_ecuador_dataset.csv")
```

```python
# quitar filas con NaN
archivo = archivo.dropna(subset=['Num. dormitorios','Num. banos','Area','Num. garages','Precio'])
```

```python
# verificar que todas las columnas sean numéricas
archivo['Num. dormitorios'] = panda.to_numeric(archivo['Num. dormitorios'], errors='coerce')
archivo['Num. banos'] = panda.to_numeric(archivo['Num. banos'], errors='coerce')
archivo['Area'] = panda.to_numeric(archivo['Area'], errors='coerce')
archivo['Num. garages'] = panda.to_numeric(archivo['Num. garages'], errors='coerce')
archivo['Precio'] = panda.to_numeric(archivo['Precio'], errors='coerce')
```

```python
# convertir a arrays
X = panda.get_dummies(
    archivo[['Provincia','Lugar','Num. dormitorios','Num. banos','Area','Num. garages']],
    drop_first=True
).astype(float).values
y = archivo['Precio'].astype(float).values

# intersección
X = numpy.c_[numpy.ones(X.shape[0]), X]

# calculo de coeficientes
coef = numpy.linalg.pinv(X) @ y

print("Coeficientes del modelo:\n\n", coef,"\n\n")

# predicciones
y_pred = X @ coef

# evaluación
mae = numpy.mean(numpy.abs(y - y_pred))
rmse = numpy.sqrt(numpy.mean((y - y_pred)**2))
ss_res = numpy.sum((y - y_pred)**2)
ss_tot = numpy.sum((y - numpy.mean(y))**2)
r2 = 1 - (ss_res/ss_tot)

print("MAE:", mae)
print("RMSE:", rmse)
print("R²:", r2)
```
